1. use IF ELSE and ELIF to write a program in python for your Report Cards

```python
roll_no = int(input("Enter the roll number"))
name = input("Enter name: ")
print("Enter 5 subject marks")
sub1 = int(input("Subject 1: "))
sub2 = int(input("Subject 2: "))
sub3 = int(input("Subject 3: "))
sub4 = int(input("Subject 4: "))
sub5 = int(input("Subject 5: "))
tot = sub1 + sub2 + sub3 + sub4 + sub5
avg = (tot) /5

if avg >= 80:
    grade = "A+"
elif avg >= 70:
    grade = "A"
elif avg >= 60:
    grade = "B+"
elif avg >= 50:
    grade = "B"
elif avg >= 40:
    grade = "C"
else:
    grade = "F"

    
if grade == "F":
    X = "FAIL"
else:
    X = "PASS"

print(roll_no,"   ",name,"    ",tot,"    ",avg,"    ",grade,"    ",X)
```

2. Use For Loop to Print Prime Numbers in between 1 to 1000 

```python
print("The prime numbers from 1 to 1000 are:")
for num in range(1,1001):
   if num > 1:
       for i in range(2,num):
           if (num % i) == 0:
               break
       else:
           print(num)
```

3. Write a program for printing the tabels from 1,10 using Nested For Loop

```python
for i in range(1, 11):
    for j in range(1, 11):
        print ('{:02}'.format(i),'X','{:02}'.format(j),'=','{:02}'.format(i*j))
    print("\n")
```
